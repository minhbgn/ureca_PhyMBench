# Training the Transition Dynamics Learning Model

## Sample Episode and Create Transition Dataset

We create the sample dataset, consist of state, action, next state, by running the envrionment and choosing random actions. We also add to this dataset the trajectories used by learnt models, so in the sample dataset there would be both random trajectories and good trajectories for the model to learn.

### Creating Model-free agent to generate good-trajectory samples

```python
#Hyperparameters used to maximize lunar lander rewards
#Used from https://huggingface.co/sb3/ppo-LunarLander-v2
hyperparameters = {
    'policy': 'MlpPolicy',
    'batch_size': 64,
    'ent_coef': 0.01,
    'gae_lambda': 0.98,
    'gamma': 0.999,
    'n_epochs': 4,
    'n_steps': 1024,
    'n_timesteps': 1000000,
    'normalize': False,  # If True, normalizes observations
    'n_envs': 16
}

log_dir = "./logs/"
```

```python
#Wrapper for Lunar Lander Environment to make it compatible with libraries expecting 4 variables for .step()
class NormalLunarLander(gym.Env):
    def __init__(self, original_env):
        super(NormalLunarLander, self).__init__()
        self.original_env = original_env
        self.action_space = self.original_env.action_space
        self.observation_space = self.original_env.observation_space
        self.state = self.original_env.reset()
        self.prev_shaping = None

    def step(self, action):
        next_state, reward, done, info, truncated = self.original_env.step(action)
        return next_state, reward, done, info

    def reset(self):
        self.state = self.original_env.reset()
        self.prev_shaping = None
        return self.state
```

```python
#Initiating the model-free agent
norm_env = NormalLunarLander(train_env)

model_free_model = PPO(
                    policy=hyperparameters['policy'],
                    env=train_env,
                    batch_size=hyperparameters['batch_size'],
                    ent_coef=hyperparameters['ent_coef'],
                    gae_lambda=hyperparameters['gae_lambda'],
                    gamma=hyperparameters['gamma'],
                    n_epochs=hyperparameters['n_epochs'],
                    n_steps=hyperparameters['n_steps'],
                    verbose=1,
                    tensorboard_log=log_dir
                  )
```

```python
#Train the model-free agent
model_free_model.learn(total_timesteps = hyperparameters['n_timesteps'])
```

```python
# Function to sample "good-trajectory" state-action-next_state samples from the learned agent
def sample_transition_data_learned(model, env, num_eps = 1000, deterministic = True):
    data = []
    for episode in range(num_eps):
        if episode % 10 == 0:
            print(f"Sampling data from episode {episode}")
        state, info = env.reset()
        done = False
        truncated = False
        while not done and not truncated:
            action, _states = model.predict(state, deterministic=deterministic)
            step_return = env.step(action)
            if len(step_return) == 4:
                next_state, reward, done, info = step_return
            else:
                next_state, reward, done, truncated, info = step_return
            if not done:
                doneType = 0
            elif env.lander.awake:
                doneType = 2
            else:
                doneType = 1
            data.append((state, action, next_state, reward, doneType))
            state = next_state
    return data
```

```python
# Function to sample random trajectories' state-action-next_state pairs
def sample_transition_data(env, num_episodes=1000):
    data = []
    for episode in range(num_episodes):
        if episode % 5 == 0:
            print(f"Sampling data from episode {episode}")
        state = env.reset()
        if type(state) is tuple:
            state = state[0]
        done = False
        truncated = False
        while not done and not truncated:
            action = env.action_space.sample()
            step_return = env.step(action)
            if len(step_return) == 4:
                next_state, reward, done, info = step_return
            else:
                next_state, reward, done, truncated, info = step_return
            if not done:
                doneType = 0
            elif env.lander.awake:
                doneType = 2
            else:
                doneType = 1
            data.append((state, action, next_state, reward, doneType))
            state = next_state
    return data
```

**Why we are generating this data:**  
- As we want to train a Neural Network model to predict next_state given the input state and action, these data including (state, action, next_state) serves as a real observed value for our training/validating/testing data set to build our neural network.

**The meaning of the tuple (state, action, next_action):**
- state: The current state that the lunar lander is in. It is an array consisting of 8 parameters: the coordinates of the lander in x & y, its linear velocities in x & y, its angle, its angular velocity, and two booleans that represent whether each leg is in contact with the ground or not. (Source: Lunar lander - Gymnasium Documentation)
- action: The action that the lunar lander takes. It can be one of 4 actions: 0: do nothing; 1: fire left orientation engine; 2: fire main engine; 3: fire right orientation engine (Source: Lunar lander - Gymnasium Documentation)
- next_state: The observed state that the lunar lander is in after taking an action. It also has 8 parameters, same as State
- (state, action) is used as our **input** for our neural network: We know exactly which state we are in, and we can choose which action to do next
- next_state is our real observed **output** after we take an action from our previous state.
- Together they form an input/output pair to be used to train our neural network model.

**The purpose of training this neural network:**
- The purpose of training this neural network is to accurately predict which state the lunar lander will be if it takes certain action
- This would aid us in choosing the action that leads to the best next state to maximize our reward

### Train the Model ###

**Build the Model**

```python
class Model(nn.Module):
    def __init__(self, in_features = 1, h1 = 16, h2 = 16, out_feature = 1):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, out_feature)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x
```

```python
model = Model(in_features = 9, h1 = 16, h2 = 16, out_feature = 8)
```

```python
model.load_state_dict(torch.load("model_lunar.pth"))
model.eval()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
X_train, X_test, y_train, y_test = getNewData(X, y)
```

```python
epochs = 500
losses = []
lossesTest = []

for i in range(epochs):
    y_pred = model.forward(X_train)
    loss = criterion(y_pred, y_train)
    losses.append(loss.detach().numpy())

    if i % 100 == 0:
        print(f'epoch: {i:2}  loss: {losses[i]}')

    with torch.no_grad():
        y_eval = model.forward(X_test)
        lossTest = criterion(y_eval, y_test)
        lossesTest.append(lossTest)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
class ModelDone(nn.Module):
    def __init__(self, in_features = 1, h1 = 10, h2 = 15, h3 = 20, h4 = 20, out_feature = 1):
        super().__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.fc3 = nn.Linear(h2, h3)
        self.fc4 = nn.Linear(h3, h4)
        self.out = nn.Linear(h4, out_feature)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = F.relu(self.fc4(x))
        x = self.out(x)
        return x

termination_model = ModelDone(in_features = 8, h1 = 10, h2 = 15, out_feature = 3)
```

```python
class CustomLunarLander(gym.Env):    
    def __init__(self, model, original_env):
        super(CustomLunarLander, self).__init__()
        self.model = model
        self.original_env = original_env
        self.action_space = self.original_env.action_space
        self.observation_space = self.original_env.observation_space
        self.state = self.original_env.reset()
        if len(self.state) == 2:
            self.state = self.state[0]
            self.info = self.state[1]
        self.prev_shaping = None
        self.terminated = False
        self.time = 0
        self.lander = self.original_env.lander

    def step(self, action):

        self.terminated = False
        self.time += 1

        state = self.state
        state_action = np.append(state, action)
        state_action_tensor = torch.FloatTensor(state_action)

        #Update state using Neural Network

        with torch.no_grad():
            state_tensor = self.model.forward(state_action_tensor)
            state = state_tensor.numpy()
            self.state = state
        
        for i in [6,7]:
          if self.state[i] > 0.5:
            self.state[i] = 1
          else:
            self.state[i] = 0

        #Original reward function from openai github:

        #Main engine firing: less reward for more engine usage
        m_power = 0.0
        if (self.original_env.continuous and action[0] > 0.0) or (
            not self.original_env.continuous and action == 2
        ):
            # Main engine
            if self.original_env.continuous:
                m_power = (np.clip(action[0], 0.0, 1.0) + 1.0) * 0.5  # 0.5..1.0
                assert m_power >= 0.5 and m_power <= 1.0
            else:
                m_power = 1.0

        #Side engine firing: less reward for more engine usage
        s_power = 0.0
        if (self.original_env.continuous and np.abs(action[1]) > 0.5) or (
            not self.original_env.continuous and action in [1, 3]
        ):
            # Orientation engines
            if self.original_env.continuous:
                direction = np.sign(action[1])
                s_power = np.clip(np.abs(action[1]), 0.5, 1.0)
                assert s_power >= 0.5 and s_power <= 1.0
            else:
                direction = action - 2
                s_power = 1.0


        reward = 0
        shaping = (
            -100 * np.sqrt(self.state[0] * self.state[0] + self.state[1] * self.state[1])
            - 100 * np.sqrt(self.state[2] * self.state[2] + self.state[3] * self.state[3])
            - 100 * abs(self.state[4])
            + 10 * self.state[6]
            + 10 * self.state[7]
        )  # And ten points for legs contact, the idea is if you
        # lose contact again after landing, you get negative reward
        if self.prev_shaping is not None:
            reward = shaping - self.prev_shaping
        self.prev_shaping = shaping

        reward -= (
            m_power * 0.30
        )  # less fuel spent is better, about -30 for heuristic landing
        reward -= s_power * 0.03

        #state[0] = (pos.x - VIEWPORT_W / SCALE / 2) / (VIEWPORT_W / SCALE / 2)
        self.original_env.lander.position.x = self.state[0] * 10 + 10

        #state[1] = (pos.y - (self.helipad_y + LEG_DOWN / SCALE)) / (VIEWPORT_H / SCALE / 2)
        self.original_env.lander.position.y = self.state[1] * 400 / 60 + 10 + self.original_env.helipad_y + 18 / 30

        #state[2] = vel.x * (VIEWPORT_W / SCALE / 2) / FPS
        self.original_env.lander.linearVelocity.x = self.state[2] * 50 / (600/30/2)

        #state[3] =  vel.y * (VIEWPORT_H / SCALE / 2) / FPS
        self.original_env.lander.linearVelocity.y = self.state[3] * 50 / (400/30/2)

        #state[4] =  self.lander.angle
        self.original_env.lander.angle = np.float64(self.state[4])

        #state[5] =  20.0 * self.lander.angularVelocity / FPS
        self.original_env.lander.angularVelocity = self.state[5] * 50 / 20

        #state[6] = 1.0 if self.legs[0].ground_contact else 0.0
        self.original_env.legs[0].ground_contact = (self.state[6] > 0.5)

        #state[7] = 1.0 if self.legs[1].ground_contact else 0.0
        self.original_env.legs[1].ground_contact = (self.state[7] > 0.5)

        #self.original_env.world.Step(1.0 / FPS, 6 * 30, 2 * 30)
        self.original_env.world.Step(1.0 / 50, 6 * 30, 2 * 30)

        terminated = False
        if self.original_env.game_over or abs(state[0]) >= 1.0:
            terminated = True
            reward = -100
        if not self.original_env.lander.awake:
            terminated = True
            reward = +100

        if self.time > 1000:
          self.terminated = True

        self.lander = self.original_env.lander

        return np.array(state, dtype=np.float32), reward, self.terminated, {}

    def reset(self):
        self.state = self.original_env.reset()
        if len(self.state) == 2:
            self.state = self.state[0]
            self.info = self.state[1]
        self.terminated = False
        self.prev_shaping = None
        self.time = 0
        return self.state

    def render(self, mode='human'):
        return self.original_env.render(mode)

#Reference: OpenAI/gym github for original environment source code
```

## Test the policy on test_env

```python
def evaluate(model, env, num_eps=1000, deterministic=True, printing = True):
    rewards = []
    for episode in range(num_eps):
        try:
            state, info = env.reset()
        except:
            state = env.reset()  # Ensure compatibility with older gym versions
            info = {}

        done, truncated = False, False
        ep_reward = 0

        while not done and not truncated:
            action, _states = model.predict(state, deterministic=deterministic)
            
            # Ensure correct unpacking of step outputs
            step_result = env.step(action)
            if len(step_result) == 4:
                next_state, reward, done, info = step_result
                truncated = False  # Assume no truncation if not returned
            else:
                next_state, reward, done, truncated, info = step_result

            ep_reward += reward
            state = next_state

        if episode % 5 == 0 and printing:
            print(f"Episode {episode}, Reward: {ep_reward}")

        rewards.append(ep_reward)

    return sum(rewards) / len(rewards)
```

Testing the model-based agent against model-free agent, by repeating these steps:

- Training both model-free and model-based agent for 10 000 steps
- Turn off the gradient and test the performance on test environment
- Stop the training once reached maximum step of 300000

```python
#Reset the agents for testing
model_free_model = PPO(
                    policy=hyperparameters['policy'],
                    env=train_env,
                    batch_size=hyperparameters['batch_size'],
                    ent_coef=hyperparameters['ent_coef'],
                    gae_lambda=hyperparameters['gae_lambda'],
                    gamma=hyperparameters['gamma'],
                    n_epochs=hyperparameters['n_epochs'],
                    n_steps=hyperparameters['n_steps'],
                    verbose=1,
                    #tensorboard_log=log_dir
                  )

rl_model = PPO(
    policy=hyperparameters['policy'],
    env=custom_env,
    batch_size=hyperparameters['batch_size'],
    ent_coef=hyperparameters['ent_coef'],
    gae_lambda=hyperparameters['gae_lambda'],
    gamma=hyperparameters['gamma'],
    n_epochs=hyperparameters['n_epochs'],
    n_steps=hyperparameters['n_steps'],
    verbose=1,
    #tensorboard_log=log_dir
    )


model_based_time = 0
model_free_time = 0
mean_rewards = []
mean_rewards_free = []
times = []
while model_based_time <= 300000 and model_free_time <= 300000:
    rl_model.learn(total_timesteps=10000)
    model_free_model.learn(total_timesteps = 10000)
    mean_reward = evaluate_policy(rl_model, test_env, n_eval_episodes=10)[0]
    mean_reward_free = evaluate_policy(model_free_model, test_env, n_eval_episodes=10)[0]
    model_based_time += 10000
    model_free_time += 10000
    mean_rewards.append(mean_reward)
    mean_rewards_free.append(mean_reward_free)
    times.append(model_free_time)
    print("Model Based: ", mean_reward,  "Model Free: ", mean_reward_free)
    print("Current time: ", model_free_time)
    if mean_reward >= 200 or mean_reward_free >= 200:
        break
        
while model_based_time <= 300000:
    rl_model.learn(total_timesteps=10000)
    mean_reward = evaluate_policy(rl_model, test_env, n_eval_episodes=10)[0]
    model_based_time += 10000
    mean_rewards.append(mean_reward)
    mean_rewards_free.append(200)
    times.append(model_based_time)
    print("mean reward: ", mean_reward, "current time: ", model_based_time)
    
while model_free_time <= 300000:
    model_free_model.learn(total_timesteps = 10000)
    mean_reward_free = evaluate_policy(model_free_model, test_env, n_eval_episodes=10)[0]
    model_free_time += 10000
    mean_rewards.append(200)
    mean_rewards_free.append(mean_reward_free)
    times.append(model_free_time)
    print("mean reward: ", mean_reward_free, "current time: ", model_free_time)

print("Model Based's maximum reward: ", model_based_time)
print("Model Free's maximum reward: ", model_free_time)
```

Plot the performance

```python
plt.plot(times, mean_rewards, label = 'Model based agent', color = 'blue')
plt.plot(times, mean_rewards_free, label = 'Model free agent', color = 'green')

plt.title('Model-based vs Model-free Test reward over time')
plt.xlabel('Time steps')
plt.ylabel('Reward')

plt.legend()
plt.show()
```

```python
def sample_transition_data_learned(model, env, num_eps = 1000, deterministic = True):
    data = []
    for episode in range(num_eps):
        if episode % 10 == 0:
            print(f"Sampling data from episode {episode}")
        state, info = env.reset()
        done = False
        truncated = False
        while not done and not truncated:
            action, _states = model.predict(state, deterministic=deterministic)
            next_step = env.step(action)
            if len(next_step) == 4:
                next_state, reward, done, info = next_step
            else:
                next_state, reward, done, info, truncated = next_step
            if not done:
                doneType = 0
            else:
                doneType = 1
            data.append((state, action, next_state, reward, doneType))
            state = next_state
    return data
```

```python
# Function to sample state-action-state pairs
def sample_transition_data(env, num_episodes=1000):
    data = {
        "state" : [],
        "action" : [],
        "next_state" : [],
        "reward" : [],
        "done" : [],
    }
    for episode in range(num_episodes):
        if episode % 1000 == 0:
            print(f"\rSampling data from episode {episode}/{num_episodes}",end='',flush=True)
        state = env.reset()
        if type(state) is tuple:
            state = state[0]
        done = False
        truncated = False
        while not done and not truncated:
            action = env.action_space.sample()
            next_step = env.step(action)
            if len(next_step) == 4:
                next_state, reward, done, info = next_step
            else:
                next_state, reward, done, info, truncated = next_step
            if not done:
                doneType = 0
            else:
                doneType = 1
            data["state"].append(state)
            data["action"].append(action)
            data["next_state"].append(next_state)
            data["reward"].append(reward)
            data["done"].append(doneType)
            state = next_state
    print(f"\rSampling data from episode {episode+1}/{num_episodes}",end='',flush=True)
        
    return data
```

```python
import pandas as pd
import numpy as np

def split_trajectories(df):
    trajectories = []
    trajectory = []

    for _, row in df.iterrows():
        trajectory.append(row.values)
        if row['done'] == 1:
            trajectories.append(np.array(trajectory))
            trajectory = []

    # Optional: if the last trajectory doesn't end with done==1
    if trajectory:
        trajectories.append(np.array(trajectory))

    return trajectories
```

**Why we are generating this data:**  
- As we want to train a Neural Network model to predict next_state given the input state and action, these data including (state, action, next_state) serves as a real observed value for our training/validating/testing data set to build our neural network.

**The meaning of the tuple (state, action, next_action):**
- state: The current state that the lunar lander is in. It is an array consisting of 8 parameters: the coordinates of the lander in x & y, its linear velocities in x & y, its angle, its angular velocity, and two booleans that represent whether each leg is in contact with the ground or not. (Source: Lunar lander - Gymnasium Documentation)
- action: The action that the lunar lander takes. It can be one of 4 actions: 0: do nothing; 1: fire left orientation engine; 2: fire main engine; 3: fire right orientation engine (Source: Lunar lander - Gymnasium Documentation)
- next_state: The observed state that the lunar lander is in after taking an action. It also has 8 parameters, same as State
- (state, action) is used as our **input** for our neural network: We know exactly which state we are in, and we can choose which action to do next
- next_state is our real observed **output** after we take an action from our previous state.
- Together they form an input/output pair to be used to train our neural network model.

**The purpose of training this neural network:**
- The purpose of training this neural network is to accurately predict which state the lunar lander will be if it takes certain action
- This would aid us in choosing the action that leads to the best next state to maximize our reward

### Train the Model ###

**Build the Model**

```python
class Model(nn.Module):
    def __init__(self, in_features = 1, h1 = 16, h2 = 16, out_feature = 1):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, out_feature)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x
```

```python
model = Model(in_features = 5, h1 = 10, h2 = 15, out_feature = 4)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
print(model.parameters)
```

```python
X_train, X_test, y_train, y_test = getNewData(X, y)
```

```python
epochs = 1500
losses = []

for i in range(epochs):
    y_pred = model.forward(X_train)
    loss = criterion(y_pred, y_train)
    losses.append(loss.detach().numpy())

    if i % 100 == 0:
        print(f'epoch: {i:2}  loss: {losses[i]}')

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
class CustomCartPole(gym.Env):
    def __init__(self, model, original_env):
        super().__init__()
        self.model = model
        self.original_env = original_env
        self.action_space = self.original_env.action_space
        self.observation_space = self.original_env.observation_space

        if len(self.original_env.reset()) == 2:
            self.state = self.original_env.reset()[0]
            print(self.state)
        else:
            self.state = self.original_env.reset()
            print(self.state)
        
        self.prev_shaping = None
        self.terminated = False
        self.steps_beyond_terminated = None
        self.time = 0
        self.time_limit = 500

    def step(self, action):

        self.terminated = False
        self.time += 1

        state = self.state
        state_action = np.append(state, action)
        state_action_tensor = torch.FloatTensor(state_action)

        #Update state using Neural Network

        with torch.no_grad():
            state_tensor = self.model.forward(state_action_tensor)
            state = state_tensor.numpy()
            self.state = state

        terminated = bool(
            state[0] < -self.original_env.x_threshold
            or state[0] > self.original_env.x_threshold
            or state[2] < -self.original_env.theta_threshold_radians
            or state[2] > self.original_env.theta_threshold_radians
        )

        if not terminated:
            reward = 1.0
        elif self.steps_beyond_terminated is None:
            # Pole just fell!
            self.steps_beyond_terminated = 0
            reward = 1.0
        else:
            if self.steps_beyond_terminated == 0:
                print(
                    "You are calling 'step()' even though this "
                    "environment has already returned terminated = True. You "
                    "should always call 'reset()' once you receive 'terminated = "
                    "True' -- any further steps are undefined behavior."
                )
            self.steps_beyond_terminated += 1
            reward = 0.0

        if self.time > self.time_limit:
            terminated = True
            reward = 1.0

        return np.array(state, dtype=np.float32), reward, terminated, {}

    def reset(self):
        if len(self.original_env.reset()) == 2:
            self.state = self.original_env.reset()[0]
        else:
            self.state = self.original_env.reset()
        self.terminated = False
        self.prev_shaping = None
        self.time = 0
        return self.state

    def render(self, mode='human'):
        return self.original_env.render(mode)

#Reference: OpenAI/gym github for original environment source code
```

```python
class SindyCartPole(gym.Env):
    def __init__(self, model, original_env):
        super().__init__()
        self.model = model
        self.original_env = original_env
        self.action_space = self.original_env.action_space
        self.observation_space = self.original_env.observation_space

        if len(self.original_env.reset()) == 2:
            self.state = self.original_env.reset()[0]
            print(self.state)
        else:
            self.state = self.original_env.reset()
            print(self.state)
        
        self.prev_shaping = None
        self.terminated = False
        self.steps_beyond_terminated = None
        self.time = 0
        self.time_limit = 500

    def step(self, action):

        if self.terminated:
            print(
                "You are calling 'step()' even though this "
                "environment has already returned terminated = True. You "
                "should always call 'reset()' once you receive 'terminated = "
                "True' -- any further steps are undefined behavior."
            )
            self.steps_beyond_terminated += 1
            reward = 0.0
            return np.array(self.state, dtype=np.float32), 0, 1, {}

        self.terminated = False
        self.time += 1

        state = self.state
        #Update state using SINDy
        state_tensor = torch.Tensor(self.model.predict(torch.Tensor(self.state).unsqueeze(0), torch.tensor(action).unsqueeze(0)))
        state = state_tensor.numpy()[0]
        self.state = state

        terminated = bool(
            state[0] < -self.original_env.x_threshold
            or state[0] > self.original_env.x_threshold
            or state[2] < -self.original_env.theta_threshold_radians
            or state[2] > self.original_env.theta_threshold_radians
        )

        self.terminated = terminated

        if not terminated:
            reward = 1.0
        elif self.steps_beyond_terminated is None:
            # Pole just fell!
            self.steps_beyond_terminated = 0
            reward = 1.0
        else:
            return np.array(state, dtype=np.float32), 0, terminated, {}

        if self.time > self.time_limit:
            terminated = True
            reward = 1.0

        return np.array(state, dtype=np.float32), reward, terminated, {}

    def reset(self):
        if len(self.original_env.reset()) == 2:
            self.state = self.original_env.reset()[0]
        else:
            self.state = self.original_env.reset()
        self.terminated = False
        self.prev_shaping = None
        self.time = 0
        return self.state

    def render(self, mode='human'):
        return self.original_env.render(mode)

#Reference: OpenAI/gym github for original environment source code
sindy_env = SindyCartPole(model, train_env)
```

## Test the policy on test_env

Report the results you get and the training/test plot. Put the results in the word file, you will reuse the reported results when writing the URECA report.

```python
def evaluate(model, env, num_eps = 1000, deterministic = True):
    rewards = []
    for episode in range(num_eps):
        if episode >= 10 and episode % 100 == 0:
            print(f"Sampling data from episode {episode}")
            #print("Episode: ", episode, "Reward: ", rewards[episode-1])
        try:
            state, info = env.reset()
        except:
            state = env.reset()
        done = False
        truncated = False
        ep_reward = []
        while not done and not truncated:
            action, _states = model.predict(state, deterministic=deterministic)
            next_step = env.step(action)
            if len(next_step) == 4:
                next_state, reward, done, info = next_step
            else:
                next_state, reward, done, truncated, info = next_step
            ep_reward.append(reward)
            state = next_state
        rewards.append(sum(ep_reward))
    return sum(rewards)/len(rewards)
```

Testing the model-based agent against model-free agent, by repeating these steps:

- Training both model-free and model-based agent for 1000 steps
- Turn off the gradient and test the performance on test environment
- Stop the training once reached maximum reward of 500

```python
#Reset the agents for testing
time_step_interval = 5000
model_free_model = PPO(
                    policy=hyperparameters['policy'],
                    env=train_env,
                    batch_size=hyperparameters['batch_size'],
                    ent_coef=hyperparameters['ent_coef'],
                    gae_lambda=hyperparameters['gae_lambda'],
                    gamma=hyperparameters['gamma'],
                    n_epochs=hyperparameters['n_epochs'],
                    n_steps=hyperparameters['n_steps'],
                    verbose=1,
                    #tensorboard_log=log_dir
                  )

rl_model = PPO(
    policy=hyperparameters['policy'],
    env=sindy_env,
    batch_size=hyperparameters['batch_size'],
    ent_coef=hyperparameters['ent_coef'],
    gae_lambda=hyperparameters['gae_lambda'],
    gamma=hyperparameters['gamma'],
    n_epochs=hyperparameters['n_epochs'],
    n_steps=hyperparameters['n_steps'],
    verbose=1,
    #tensorboard_log=log_dir
    )


model_based_time = 0
model_free_time = 0
mean_rewards = []
mean_rewards_free = []
times = []
while True:
    rl_model.learn(total_timesteps=time_step_interval)
    model_free_model.learn(total_timesteps = time_step_interval)
    mean_reward = evaluate(rl_model, test_env, num_eps=20)
    mean_reward_free = evaluate(model_free_model, test_env, num_eps=20)
    model_based_time += time_step_interval
    model_free_time += time_step_interval
    mean_rewards.append(mean_reward)
    mean_rewards_free.append(mean_reward_free)
    times.append(model_free_time)
    print("Model based: ", mean_reward, "Model free: ", mean_reward_free)
    if mean_reward >= 500 or mean_reward_free >= 500:
        break
        
while mean_reward < 500:
    rl_model.learn(total_timesteps=time_step_interval)
    mean_reward = evaluate(rl_model, test_env, num_eps=20)
    model_based_time += time_step_interval
    mean_rewards.append(mean_reward)
    mean_rewards_free.append(500)
    times.append(model_based_time)
    print("mean reward: ", mean_reward, "current time: ", model_based_time)
    
while mean_reward_free < 500:
    model_free_model.learn(total_timesteps = time_step_interval)
    mean_reward_free = evaluate(model_free_model, test_env, num_eps = 20)
    model_free_time += time_step_interval
    mean_rewards.append(500)
    mean_rewards_free.append(mean_reward_free)
    times.append(model_free_time)
    print("mean reward: ", mean_reward_free, "current time: ", model_free_time)

print("Model Based time to get maximum reward: ", model_based_time)
print("Model Free time to get maximum reward: ", model_free_time)
```

```python
plt.plot(times, mean_rewards, label = 'Model based SINDy agent', color = 'blue')
plt.plot(times, mean_rewards_free, label = 'Model free agent', color = 'green')

plt.title('Model-based vs Model-free Maximum reward time')
plt.xlabel('Time steps')
plt.ylabel('Reward')

plt.legend()
plt.show()
```
